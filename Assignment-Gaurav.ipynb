```python
xl2=pd.read_excel("D:\Excel sheet.xlsx", usecols=['Name', 'Market Cap'])
xl2.head()
```

```python
xl2['Market Cap']=xl2['Market Cap']/10
xl2.head()
```

```python
def m_cap(cell):
    if cell>20000:
        return "Large Cap"
    elif cell<20000 and cell>5000:
        return "Mid Cap"
    elif cell<5000:
        return "Small Cap"

xl2['Cap']=xl2['Market Cap'].apply(m_cap)
xl2.head()
```

```python
xl2.groupby('Cap').size()
```

```python
xl2.groupby(['Cap']).size().plot(kind='pie', y='Cap')
```
